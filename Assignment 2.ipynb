```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
columns = [
    'mpg', 'cylinders', 'displacement', 'horsepower',
    'weight', 'acceleration', 'model_year',
    'origin', 'car_name'
]
```

```python
!1s"/content/drive/MyDrive/Colab Notebook"
```

```python
import zipfile
zip_path="/content/drive/MyDrive/auto+mpg (1).zip"
with zipfile.ZipFile(zip_path,'r') as zip_ref:
  zip_ref.extractall("/content/auto_mpg")
```

```python
df=pd.read_csv(
    "/content/auto_mpg/auto-mpg.data",
    delim_whitespace=True,
    names=columns,
    na_values='?'
)
```

```python
df.dropna(inplace=True)
```

```python
print(df.head())
```

```python
print("\nDescriptive Statistics")
print(df.describe())
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['weight'], df['mpg'], color='blue')
plt.title("Weight vs Mileage")
plt.xlabel("Car Weight")
plt.ylabel("Mileage (MPG)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['mpg'], bins=10, edgecolor='black')
plt.title("Histogram of Mileage")
plt.xlabel("Mileage (MPG)")
plt.ylabel("Frequency")
plt.show()
```

```python
x = df[['displacement']]
y = df['mpg']
```

```python
model = LinearRegression()
model.fit(x, y)

predictions = model.predict(x)

print("\nIntercept:", model.intercept_)
print("Coefficient:", model.coef_[0])
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['displacement'], y, color='blue', label='Actual Data')
plt.plot(df['displacement'], predictions, color='red', linewidth=2, label='Regression Line')
plt.title("Engine Size vs Mileage")
plt.xlabel("Engine Size (Displacement)")
plt.ylabel("Mileage (MPG)")
plt.legend()
plt.grid(True)
plt.show()
```
